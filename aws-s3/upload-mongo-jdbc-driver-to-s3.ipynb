If you don't have an S3 bucket yet, you can create one using the AWS CLI or via the AWS Management Console. Below are the steps to create a new S3 bucket and upload the JDBC driver:

### Step 0: Download MongoDB JDBC Driver:

Available at: https://www.mongodb.com/try/download/odbc-driver

### Step 1: Create an S3 Bucket

#### **Using AWS CLI**:
To create a new S3 bucket using the AWS CLI, use the following command:

```bash
aws s3 mb s3://your-bucket-name
```

- Replace `your-bucket-name` with the name you want for your S3 bucket.
- **Bucket Name Rules**:
  - Must be globally unique.
  - Can contain only lowercase letters, numbers, hyphens, and periods.
  - Cannot start or end with a hyphen or period.

For example:
```bash
aws s3 mb s3://my-glue-jdbc-bucket
```

This command creates a new bucket named `my-glue-jdbc-bucket` in your default AWS region.

#### **Via AWS Management Console**:
1. Go to the [Amazon S3 Console](https://s3.console.aws.amazon.com/s3).
2. Click **Create bucket**.
3. Enter a unique bucket name (e.g., `my-glue-jdbc-bucket`).
4. Choose a region for your bucket (you can keep the default).
5. Leave the other settings as default and click **Create**.

### Step 2: Upload the JDBC Driver to S3

Once the bucket is created, you can upload the JDBC driver to it.

#### **Using AWS CLI**:
To upload the JDBC driver, use the `aws s3 cp` command as mentioned before.

For example:
```bash
aws s3 cp mongodb-jdbc-2.2.1-all.jar s3://my-glue-jdbc-bucket/glue-jdbc-drivers/
```

This command will upload the `mongodb-jdbc-2.2.1-all.jar` file to the `glue-jdbc-drivers/` directory inside your S3 bucket `my-glue-jdbc-bucket`.

#### **Via AWS Management Console**:
1. Open the [S3 Console](https://s3.console.aws.amazon.com/s3).
2. Select your newly created bucket.
3. Click **Upload**.
4. Choose **Add files**, then select the `mongodb-jdbc-2.2.1-all.jar` file from your local machine.
5. Click **Upload**.

### Step 3: Verify the Upload

To verify the upload, you can either:
- Check the contents of your S3 bucket from the **S3 Console**.
- Use the AWS CLI to list the files in your bucket:
  ```bash
  aws s3 ls s3://my-glue-jdbc-bucket/glue-jdbc-drivers/
  ```

Once you've uploaded the JDBC driver to the S3 bucket, you can reference it in AWS Glue as described earlier.
